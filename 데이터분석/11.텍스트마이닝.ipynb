# 텍스트 관련 분야

- **텍스트 마이닝**
  - 문자로된 데이터에서 가치있는 정보를 얻어내는 과정
  - 주제
    - 워드 카운트, 핵심 키워드 추출, 워드 클라우드(시각화), 유사도 계산통한 챗봇, ... 
  - **텍스트를 어떻게 처리해서 수치화**
    - **토큰화(tokeniztion)**
      - 주어진 코퍼스(ex) 말뭉치)에서 토큰(token)단위로 분절(분해한다)하는 과정(분절의 결과는 토큰으로 쪼개진다)으로 이 과정전체를 토큰화라고 표현
    - 정제(cleaning)
    - 정규화(normalizaion)
    - 사전화 -> 백터화
  - 데이터 분석 분야

- NLP
  - 자연어처리(Netural Language Processing)
  - 자연어를 컴퓨터가 인지하는 언어로(0.1) 변환하여 인지하게 하고 처리하는것
  - 머신러닝/딥러닝 분야에서 처리하는 게 일반적
  - 텍스트 자체가 데이터이면서 정답이다
  - 주제
    - 기계번역, 긍정/부정등 예측, 키워드 추출, 요약, 글생성, 챗봇, ...
  - 텍스트를 어떻게 처리해서 수치화
    - 말뭉치 -> 분절화(토큰화) -> 사전화 -> 백터화 -> 원-핫인코딩(선택적) -> 압축(임베딩) -> 정규화(필요시)

## 구동 테스트

```python
ko_sentence = '''
9일 오전 아시아 주요 증시는 일본 시장이 '성인의 날'로 휴장한 가운데 오름세를 나타내고 있다. 전날부터 시작된 중국 국경 재개방으로 인한 관광산업 회복과 금융당국의 기술기업 규제 완화 기대가 중화권 시장 참여자들의 투자심리를 자극했다. 일본 증시는 1월 둘째 주 월요일을 '성인의 날' 연휴로 지정하고 하루 휴장한다.
'''.strip()
ko_sentence
```

```python
# mecab 테스트
from konlpy.tag import Mecab
```

```python
tokenizer = Mecab()
```

```python
print( tokenizer.morphs( ko_sentence )  )
# 품사별로 토큰을 분해해서 리스트로 리턴
# 문장의 개수, 구성에 상관없이 토큰화 진행
```

```python
print( tokenizer.pos( ko_sentence ) )
# (토큰, 품사코드(태그)) 추출, 토큰의 성분을 알수 잇다. 품사태깅
```

```python
# 명사만 추출, 키워드 추출시 사용 가능
print( tokenizer.nouns( ko_sentence ) )
```

```python
# kss 확인
import kss

# 리스트 형태로 문장 제공
print( kss.split_sentences( ko_sentence ) )
```

```python
# 품사 태킹 mecab.pos()
print( kss.split_morphemes( ko_sentence ) )
```

## 워드클라우드

- 긍정 리뷰에서 많이 나오는 토큰을 워드 클라우드 형식으로 시각화 한다

```python
!pip install wordcloud
```

```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np

# 이미지 위에 워드클라우드를 그리기 위함
from PIL import Image
```

```python
# 이미지 로드 -> 이미지를 배열로 추출
heart_mask = np.array( Image.open('/content/heart.jpg') )
```

```python
# 기본 그리기 (워드클라우드 베이스 그림 체크)
plt.figure( figsize=(7,7) )
plt.imshow( heart_mask, cmap=plt.cm.gray )
plt.axis('off')
plt.show()
# 이미지 정상적으로 보이는것 확인
```

```python
# 불용어 -> 워드 클라우드 상에서 배제할 텍스트 추가
stopwords = set(STOPWORDS)

# 추가
# 불용어로 정의한 토큰들이 없다 절차는 생략
stopwords.add('제외한단어')
```

```python
# 워드 클라우드 생성
wc = WordCloud( 
  font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf',  #폰트지정 
  max_words = 300, # 빈도 서열 300위까지 클라우드에 표현
  stopwords = stopwords,   # 불용어 추가
  mask = heart_mask,       # 이부분이 생략되면 사각형 안에서 표시
  background_color='white' #  마스크 베이스를 흰색 세팅
)
```

```python
# 데이터 주입 -> 워드 카우트 진행 -> 빈도 서열 준비
wc_gen = wc.generate( review_posi_text )
```

```python
# 빈도 서열 확인
# 최대 빈도를 1.0으로 맞추고, 나머지 빈도는 서열화 ~ => max값기준정규화 처리 
wc_gen.words_
```

```python
# 실습 : wc_gen.words_ 을 이용하여 바차트 그리시오
# X는 토큰, Y는 빈도, 상위 20개만 대상 -> 5분
wc_df = pd.DataFrame()
wc_df['word'] = [ k for k in wc.words_.keys() ]
wc_df['freq'] = [ v for v in wc.words_.values() ]
wc_df.head(2)
```

```python
# 원본데이터가 이미 정렬이 되었다고 하지만, 다시한번 정확하게 진행
wc_df_sorted = wc_df.sort_values(by='freq', ascending=False)
wc_df_sorted.head(2)
```

```python
wc_df_sorted[ wc_df_sorted.index < 20].plot( kind='bar', x='word', rot=45)
# 데이터를 살펴보니, 무의미한 의성어, 품사 등등이 보인다 => 정제후 다시
```

```python
plt.figure( figsize=(7,7) )
plt.imshow( wc_gen )
plt.axis('off')
plt.show()
# 텍스트에서 명사위주로 남기면 더욱 유의미가 있지 않을까 => 
# 실습(16시이후)#1 : 코퍼스 => 품사태깅 => 명사만 남기고 => 다시 코퍼스 구성 => 워드클라우드 생성
# 실습(16시이후)#2 : 부정리뷰만 모아서 다시 진행
# 실습(16시이후)#3 : 긍정 리뷰 상위 키워드, 부정 리뷰 상위 키워드 비교 (탑 20위권)
```

## 텍스트 유사도를 계산하여 챗봇 구현

- 유사도 계산
  - 코사인 유사도 사용
  - 텍스트 -> 수치화
    - 문장1개 -> 백터화 [ 1, 0, 3, 4, 5, 3, 0 , 0 ..]
    - A문장과 가장 가까운 B문장을 챗봇시트 질문목록에서 찾아낼수 있다
      - 2개의 문장의 거리가 가깝다 => 의미가 유사하다

### 말뭉치 -> 사전화 -> 백터화

- 말뭉치(코퍼스)
  - 텍스트 덩어리 
  - 2차원 구조에 문자열로 구성된 문장의 배열일수도 있다

### 카운트 기반 단어 표현(Count based word Representation)


- 국소 표현 방법(local Representation)
  - 단어를 새로 많나면 번호(특정값)를 부여한다 
  - cat, dog, love 라는 단어(토큰)를 처음 만나면(사전화 작업할때), 1번, 2번, 3번 부여했다(맵핑) 
  - 뉘앙스가 없다
  - 이산표현
  - 예)
    - Bag of Words
    - DTM
    - **TF-IDF(이 방식을 사용하겟다)**

- 분산 표현 방법(Distributed Representation)
  - 단어를 표현할때, 주변을 참고하여 단어를 표현한다
  - 뉘앙스가 있다
  - dog이라는 단어 주변에 cute, lovely등등 단어들이 자주 등장한다면, dog, cute, lovely이런단어들은 비슷한 느낌을 가진다 명시(가중치를 부여, 비중을 부여)
  - 연속형으로 표현된다
  - 예)
    - **Word2Vec**
    - FastTest
    - Glove

### TF-IDF 직접만들기

- 소량의 텍스트로 구현
- 대량의 데이터는 라이브러리 사용 (머신러닝 모듈 활용)

#### 개별 함수 생성

- **TF-IDF**(단어 빈도-역문서빈도)
  - 2개의 정보를 혼합해서 단어를 수치화 하여 사전화 처리 했다
  - **T**erm **F**requency - **I**nverse **D**ocument **F**requency
  - 정의
    - 단어의 빈도(전체 텍스트에서 단어가 몇번 나오는가)와 역외빈도(여러문장에서 이 단어가 몇번 나오는가)를 연산했다
    - 국소 표현 방법(local Representation)중에서는 가장 발전한 방식이다
    - **각 단어들마다 중요한 정도를 가중치로 주는 방식을 채택**
  - 활용
    - 문서의 유사도 검사 
    - 검색시스템 검색결과의 중요도 표시
    - 문서내 특정 단어 중요도 처리


```python
import pandas as pd
from math import log  # IDF 계산용
```

```python
# 말뭉치 => 사전화 구현 (Mecab과 같은 토큰화 도구 미사용:컨셉)
docs = [
    '먹고 싶은 사과',
    '먹고 싶은 돈까스',
    '맛있는 치즈 돈까스 돈까스',
    '저는 과일이 좋아요'
]
# 공백 단위로 쪼개서 중복제거(사전에는 1개의 단어만 등록), 정렬
vocab = list(set(w for doc in docs for w in doc.split()))
vocab.sort()
vocab
# 현재 주어진 코퍼스 기준으로 사전을 만들었다
```

```python
# 계산식

# 1. 문장의 수
N = len( docs )
N
```

```python
# 2. TF-IDF에서 TF를 구하는 함수
# 문장내에서 사전에 등록한 토큰의 카운트 계산
def tf(t, d):
  '''
    t : 토큰
    d : 문장
  '''
  return d.count(t)
# 아래 문장에서 돈까스의 카운트를 계산하시오
'맛있는 치즈 돈까스 돈까스'.count('돈까스')

# 테스트
tf( '돈까스', '맛있는 치즈 돈까스 돈까스')
```

```python
# 조건식애 in을 사용하면 내부에 해당값이 있는지 체크
'돈까스' in '맛있는 치즈 돈까스 돈까스'
```

```python
# 3. IDF 함수 구현
# 문장에서 해당 토큰이 발견되면 +1 => 
# 모든 문장을 순회하면 최종 리턴(약간의 연산)=>가중치=>이방식을 고안한 쪽에 공식을 사용
def idf(t):
  # 전역변수로 정의된 docs를 그냥 사용하겟다
  cnt = 0
  for doc in docs:
    cnt += t in doc
  #print( cnt )
  # 가중치 계산 => IDF 식에 의해서 구현 (차후 체크 -> 논문이나 해당 이론 체크, 도출)
  return log( N/(cnt+1) )

# 실습 4 : vocab에 있는 모든 토큰의 가중치를 계산해서 합산하시오
idf('먹고'), idf('돈까스')
```

```python
# 4. tf-idf 값 계산 함수
def tf_idf( t, d ):
  return tf(t, d) * idf(t)
```

### 유사도 계산식 구현

- cos유사도
- 직접 구현 가능(numpy)
- 제공( scipy )

```python
import numpy as np

def cos_sim(AVec, BVec):
  '''
    - 코사인 유사도 구현
    - Args
      - AVec : 문장을 수치화한 백터 : 사용자가 입력하는 질문
      - BVec : 문장을 수치화한 백터 : 챗봇 시트에 등록된 질문 목록 : Q2
    - Returns
      - 유사도값 -> 차후 해단값이 가장 best 질문을 찾는데 활용된다(최고값)
  '''
  return np.dot(AVec, BVec) / ( np.linalg.norm(AVec) * np.linalg.norm(BVec) ) 
```

### 질문 건당 가장 최상의 유사한 챗봇시트 질문지 찾은후 답변 리턴

```python
# 사전에 챗붓 시트 Q2를 백터로 계산해 둔다
chatbot_df['Q2_vec'] = chatbot_df.Q2.apply( lambda x: vectorizer.transform([x]).toarray()[0] )
chatbot_df.head(1)
```

```python
chatbot_df.Q2_vec[0].shape
```

```python
def check_answer_sim( user_sentnece='' ):
  if not user_sentnece:
    return '정확하게 입력후 문의하세요'
  # 1. 최초는 파생변수가 생성(없다면) -> 이후에는 0으로 초기화 
  chatbot_df['score'] = 0

  # 2. 사용자의 입력을 백터화
  # 문장 => Mecab 분절 => 문장으로 재구성 => 백터화 => 배열(2차원, matrix)추출 => 백터(1차원) 추출
  user_sentnece_vec = vectorizer.transform( [ my_sentence_change(user_sentnece) ] ).toarray()[0]    

  # 3. 사용자의 입력 백터와 챗봇 시트.Q2_vec 백터간에 유사도를 계산하여 score에 담아야한다
  # 챗봇 시트.Q2 사전에 미리 백터화 해둬야 한다
  # (변경되지 않는 내용을 매번 계산하는것은 비효울-> 응답이 늦음)
  chatbot_df['score'] = chatbot_df.Q2_vec.apply( lambda x: cos_sim(user_sentnece_vec, x) )
  #print(chatbot_df['score'])

  # 4. 이중 최고 스코어를 받은 인덱스(가장 유사한 질문지를 찾았다)를 찾아서 => idxmax()
  #    해당 질문과 매칭된 답변을 리턴한다
  return chatbot_df.loc[ chatbot_df['score'].idxmax() ]['A']
```
