# 메타 문자

## (1) |

- 의미  or

```python
# re.match( 정규식, 데이터 ) => True/False  리턴
# 정규식 패턴에 데이터가 합당한지 체크
if re.match( 'a|b|c', 'a' ): print('1. 매치된다')
if re.match( 'a|b|c', 'b' ): print('2. 매치된다')
if re.match( 'a|b|c', 'c' ): print('3. 매치된다')
```

```python
if re.match( 'a|b|c', 'd' ): print('4. 매치된다')
if re.match( 'a|b|c', 'ad' ): print('5. 매치된다')
# 첫글자가 a 혹은 b 혹은 c 인경우에만 매칭-> 첫글자가 일치해야한다 -> match
if re.match( 'a|b|c', 'da' ): print('6. 매치된다')
```

## (2) []

- 문자 클레스
- [ 와 ] 사이에는 어떤 문자도 들어갈 수 있다
- []는 한개의 문자를 표현한다
- match(시작단어 일치), search(전체중에일치되는게 잇다면 OK)
- 만약 한개 문자를 여러개로 표현하고 싶다면
  - +(1 ~ 무한대) or *(0 ~ 무한대)를 뒤에 붙이면 된다

```python
# 문자 1개가 올수 있고, a or b or c중에 하나만 올수 있다
if re.match( '[abc]', 'a' ): print('1. 매치된다')
if re.match( '[abc]', 'efz' ): print('2. 매치된다')
# 첫글자가 a이므로 ok
if re.match( '[abc]', 'abd' ): print('3. 매치된다')
if re.match( '[abc]', 'b56' ): print('4. 매치된다')

# 첫글자에 상관없이 있기만 하면 ok
if re.search( '[abc]', '56b' ): print('5. 매치된다')
```

## (3) -

- [] 문자클레스 안에서는 두개의 문자사이에서 범위롤 표현
- from ~ to

```python
if re.match( '[0123456789]', '1' ): print('1. 매치된다')
# 0부터 9까지 숫자가 첫글자가 된다
if re.match( '[0-9]', '1' ): print('2. 매치된다')
```

```python
# 소문자로 시작하면 OK
if re.match( '[a-z]', '1' ): print('1. 매치된다')

# 소문자대문자로 시작하면 ok
if re.match( '[a-zA-Z]', 'v' ): print('2. 매치된다')
```

```python
# 숫자, 소문자만, 
# 첫글자가 대문자라서 해당 않됨
if re.match( '[0-9a-z]', 'Aa' ): print('1. 매치된다')
if re.search( '[0-9a-z]', 'Aa' ): print('2. 매치된다')
```

```python
# 한글 완성형 (utf-8)
# 자음시작-자음끝, 모음시작-모음끝, 완성글자시작-완성글자끝
# 자음시작 - 완성글자끝 => 한글범위(utf-8) 기준
if re.match( '[ㄱ-힣]', '가123ABC' ): print('1. 매치된다')
# 첫글자가 한글이므로 매칭
```

## (4) .

- 줄바꿈 기호 (\n)을 제외하고 모든 문자를 대변
- [ ] 이 안에서는 .은 .이라는 문자 하나이다
- **사용위치에 따라 해석이 다름**

```python
if re.match( '[a.b]', 'a' ): print('1. 매치된다')
if re.match( '[a.b]', 'b' ): print('2. 매치된다')
# 문자클레스 내에서는 .는 .다
if re.match( '[a.b]', 'c' ): print('3. 매치된다')
```

```python
if re.match( '[a.b]', 'ab' ): print('1. 매치된다')
if re.match( '[a.b]', 'a1b' ): print('2. 매치된다')
if re.match( '[a.b]', 'abc' ): print('3. 매치된다')
if re.match( '[a.b]', 'bcd' ): print('4. 매치된다')
if re.match( '[a.b]', 'cab' ): print('5. 매치된다')

if re.match( '[a.b]', '.ab' ): print('6. 매치된다')
```

```python
# 않되는이유 : 시작문자가 a문자b 이렇게 되어야 한다
# []를 제외하는 순간 해석이 변경된다
if re.match( 'a.b', 'a' ): print('1. 매치된다')
if re.match( 'a.b', 'abc' ): print('2. 매치된다')

if re.match( 'a.b', 'a1b' ): print('3. 매치된다')
if re.match( 'a.b', 'a2be' ): print('4. 매치된다')
if re.match( 'a.b', 'a\nb' ): print('5. 매치된다') # \n는 제외 
```

## (5) ^

- 일반 문자열에서는 문자열의 맨 처음과 일치함을 의미
- 문자열 클레스에서는 해당 문자는 배제(제외)를 가짐
- re.MUTLLINE과 같이 사용하면 각 라인의 처음 문자를 검사해서 일치함을 조사 (컴파일에서확인)

## (6) $

- 문자열의 맨끝과 일치

```python
# 숫자나 .으로 시작한다
if re.match( '^[0-9\.]', '1' ): print('1. 매치된다')
```

```python
# 숫자나 .으로 시작하고, 숫자나 .으로 끝난다
if re.match( '^[0-9\.]$', '.' ): print('1. 매치된다')
```

```python
# 매칭않됨 => [ ]는 문자 1개를 의미, 그런데 문자가 2개이다 그래서 않됨 -> *, +를 붙이면 해결된다
if re.match( '^[0-9\.]$', '11' ): print('1. 매치된다')
if re.match( '^[0-9\.]$', '1.' ): print('2. 매치된다')
if re.match( '^[0-9\.]$', '.1' ): print('3. 매치된다')
if re.match( '^[0-9\.]$', '..' ): print('4. 매치된다')
```

```python
# 끝패턴 연습
if re.match( '^[0-9]+[a-z]+$', '1' ): print('1. 매치된다')
if re.match( '^[0-9]+[a-z]+$', '11' ): print('2. 매치된다')
if re.match( '^[0-9]+[a-z]+$', '111111A' ): print('3. 매치된다')
# 숫자가 1~ 무한대로 시작하고, a-z가 1 ~ 무한대로 나와서 종료
if re.match( '^[0-9]+[a-z]+$', '1111111a' ): print('4. 매치된다')
```

## (7) *, +

- * : 이것을 붙인 앞문자은 0 ~ 무한대로 반복 가능
- + : 이것을 붙인 앞문자은 1 ~ 무한대로 반복 가능
  - 문자 클레스 내부에서는 단순 +가(이스케이프문자 권장) 될수도 있고, 반복의 의미가 될수도 있다 => 잘사용요만


```python
if re.match( '^[0-9\.]+$', '11' ): print('1. 매치된다')
if re.match( '^[0-9\.]+$', '1.' ): print('2. 매치된다')
if re.match( '^[0-9\.]*$', '.1' ): print('3. 매치된다')
if re.match( '^[0-9\.]*$', '..' ): print('4. 매치된다')
```

```python
# 숫자를 제외하고 어떤 문자던 시작문자가 될수 잇다
if re.match( '[^0-9]', '1' ): print('1. 매치된다')
if re.match( '[^0-9]', '가' ): print('2. 매치된다')
if re.match( '[^0-9]', 'a' ): print('3. 매치된다')
if re.match( '[^0-9]', 'abcd' ): print('4. 매치된다')
```

```python
# 반복 연속
# *바로 앞에 문자만 따진다 0 ~ 무한대
if re.match( 'seo*ul', 'seul' ): print('1. 매치된다')
if re.match( 'seo*ul', 'seoul' ): print('2. 매치된다')
if re.match( 'seo*ul', 'seoooul' ): print('3. 매치된다')
```

```python
# + 바로 앞에 문자만 따진다 1 ~ 무한대
if re.match( 'seo+ul', 'seul' ): print('1. 매치된다') # 반드시 1번은 나와야 함
if re.match( 'seo+ul', 'seoul' ): print('2. 매치된다')
if re.match( 'seo+ul', 'seoooul' ): print('3. 매치된다')
```

(8) {}

- 지정된 횟수 반복
- {2} : 2회 반복
- {2,3} : 2~3회까지 반복 가능

```python
if re.match( 'seo{2}ul', 'seul' ): print('1. 매치된다')
if re.match( 'seo{2}ul', 'seoul' ): print('2. 매치된다')
if re.match( 'seo{2}ul', 'seooul' ): print('3. 매치된다')
```

```python
# {} 는 무의미, 그냥 문자 표기한것 => 의도와 다르게 작동
# 문자클렉스 않에서는 일단 무의미
if re.match( '[seo{2}ul]', 'seul' ): print('1. 매치된다')
```

```python
if re.match( 'seo{2,3}ul', 'seul' ): print('1. 매치된다')
if re.match( 'seo{2,3}ul', 'seoul' ): print('2. 매치된다')
# 2회 반복 ~ 
if re.match( 'seo{2,3}ul', 'seooul' ): print('3. 매치된다')
# 3회 반복만 해당됨
if re.match( 'seo{2,3}ul', 'seoooul' ): print('4. 매치된다')
if re.match( 'seo{2,3}ul', 'seooooul' ): print('5. 매치된다')
```

## (10) 메타문자 > 이스케이프 문자

- \d <-> \D
- \s <-> \S
- \w <-> \W
- \b <-> \B
- 기타
  - 일반적인 특수문자를 . => \.  표기
  - 화이트스페이스 \t, \n, \v, \r



```python
# \d 숫자 => [0-9]
if re.match( '[\d]' , '1'): print('1 매치된다')
if re.match( '[\d]' , 'a'): print('2 매치된다')
```

```python
# ^ => xor 연산 : 일반코드상에서 의미
# \D 숫자 => [^0-9]
if re.match( '[\D]' , '1'): print('1 매치된다')
if re.match( '[\D]' , 'a'): print('2 매치된다')
```

```python
# \s => 화이트스페이스 : \n \t \r \v
if re.match( '[\s]' , '\n'): print('1 매치된다')
if re.match( '[\s]' , 'a'): print('2 매치된다')
```

```python
# \S => 화이트스페이스가 아닌 문자
if re.match( '[\S]' , '\n'): print('1 매치된다')
if re.match( '[\S]' , 'a'): print('2 매치된다')
```

```python
# \w -> 문자, 숫자 매치
if re.match( '[\w]' , 'a'): print('1 매치된다')
if re.match( '[\w]' , '1'): print('2 매치된다')
if re.match( '[\w]' , '가'): print('3 매치된다')
if re.match( '[\w]' , 'A'): print('4 매치된다')
if re.match( '[\w]' , '@'): print('5 매치된다')
```

```python
# \W => 문자와 숫자가 아닌 특수문자, 화이트스페이스
if re.match( '[\W]' , 'a'): print('1 매치된다')
if re.match( '[\W]' , '@'): print('2 매치된다')
if re.match( '[\W]' , '\t'): print('3 매치된다')
```

# 함수
  

- 해당 문자열(정규식표현)이 존재하는지 중요
  - **match()**
    - 문자열의 처음(시작)부터 정규식과 매칭되는 조사
    - 매치되면 매치된내용을 응답(re.Match object)
  - **search()**
    - 문자열의 전체를 검색하여, 매칭되는 조사
    - 매치되면 매치된내용을 응답(re.Match object)

- 정규식에 해당되는 내용을 가져와라 
  - findall()
    - 정규식에 매칭되는 모든 문자열을 list로 리턴
  - finditer()
    - 정규식에 매칭되는 모든 문자열을 iterator 리턴

- **sub()**
  - 정제(클리닝) 관점
  - 정규식에 매칭이 되는 부분은 특정 문자열로 대체

```python
p = re.compile('[0-9]+')
```

```python
# 시작부터 정규식에 매칭되지 않아서 None
res = p.match( 'abcd5' )
print(res)
```

```python
res = p.match( '5acvb' )
if res: # 매칭된다
  print( res.group(), '매치되는 데이터가 있다' ) # 매칭되는 데이터는 group() 추출
```

```python
res = p.search( 'abcd5 5skdn 1234 sdfjw434' )
print(res)
```

```python
# 문장에서 최초 발견된면 탐색 끝
if res:
  print( res.group() ) # 이것을 사용
  print( res.groups() )
  print( res.group(0) ) # 사실상 무의미
```

```python
# 정규식 새로 정의
p = re.compile('[a-z]+') # 소문자만 1~무한대
```

```python
# 문자열에서 소문자로만 이어지 문자열을 다 찾아서 리스트로 리턴
res = p.findall( 'abcd5 5skdn 1234 sdfjw434 12abc34 ad12fh' )
if res:
  print( res )
```

```python
# 검색된 부분이 문자열내에 어떤 위치에 있었는지도 제공
res = p.finditer( 'abcd5 5skdn 1234 sdfjw434 12abc34 ad12fh' )
for w in res:
  print( w, w.group(), w.start(), w.end(), w.span() )
```

# 그룹핑

- 정보를 그룹화하여서 간편하게 추출 가능
- res.group(0)
  - 매칭되는 모든 문자열을 출력
- res.group(1 or 2 or 3..) 
  - 첫(두 or 세..)번째 그룹의 문자열

```python
text = '동국대역 010-1234-5678 급매' # 보증금 없음 월세 10만원'
text
```

```python
p = re.compile( r'\w+\s+\d+[-]\d+[-]\d+\s+\w+' )
```

```python
m = p.search( text )
m
```

```python
m.group(0)
```

- 전화번호를 뽑는데 정규식이 전혀 도움이 않되었다(하나마다)
- 전화번호에 대한, 특정 타겟에 대한 그룹화 처리가 없었다 -> 적용하면 내용별, 번호별, 번호 성분별 추출 가능하다

```python
# 적용  -> ()
p = re.compile( r'(\w+)\s+(\d+[-]\d+[-]\d+)\s+(\w+)' )
m = p.search( text )
m
```

```python
# 아래 두개는 같은 내용
m.group(), m.group(0)
```

```python
m.group(1)
```

```python
m.group(2)
```

```python
m.group(3)
```

```python
# 전화번호 3-4-4도 각각 그룹화하여서 추출하시오
p = re.compile( r'(\w+)\s+((\d+)[-](\d+)[-](\d+))\s+(\w+)' )
m = p.search( text )
```

```python
for n in m.groups():
  print( n )
```

- 그룹핑된 문자열의 재참조
- 그룹핑된 그룹은 
  - 1(첫번째그룹), 2, 3 번 인덱스로 부른다
  - \1, \2, \3, ...

```python
# 자기 참조식 -> 정규식 패턴도 일치, 그 값도 일치해야 한다 => 매칭된다
p = re.compile( r'(\b\w+)\s+\1'  )
m = p.search( 'python numpy pandas pandas tensorflow' )
m.groups(), m.group(), m.group(1)
```

- 그룹핑 네이밍 (이름을 설정해서 참조)
- 그룹에서 값 추출할때 편하다
- (?P<이름>정규식)

```python
p = re.compile( r'(?P<AREA>\w+)\s+((?P<HP1>\d+)[-](?P<HP2>\d+)[-](?P<HP3>\d+))\s+(\w+)' )
m = p.search( text )
```

```python
# 추출한 데이터의 의미를 이해하고 추출 가능, 순서는 몰라도 된다
m.group('AREA'), m.group('HP1'), m.group('HP2'), m.group('HP3')
```

- 자기참조에 이름붙이기

```python
# (?P=이름)
# \1 => (?P=LAN)
p = re.compile( r'(?P<LAN>\b\w+)\s+(?P=LAN)'  )
m = p.search( 'python numpy pandas pandas tensorflow' )
m.groups(), m.group(), m.group(1)
```

## DOTALL, S

- 모든문자, \n포함

```python
p = re.compile( 'a.b' )
print( p.match('a|b') )
print( p.match('a\nb') )
```

```python
p = re.compile( 'a.b', re.DOTALL )
print( p.match('a|b') )
print( p.match('a\nb') ) #  줄바꿈도 처리된다
```

```python
p = re.compile( 'a.b', re.S )
print( p.match('a|b') )
print( p.match('a\nb') ) #  줄바꿈도 처리된다
```

## IGNOGRECASE, I

- 대소문자 구분없이 매칭

```python
p = re.compile( '[a-z]' )
print( p.match('python').group() )
print( p.match('Python') )
```

```python
p = re.compile( '[a-z]', re.I )
print( p.match('python').group() )
print( p.match('Python').group() )
```

# 전방탐색(Lookahead Asserions)

- positive (긍정)
  - 긍정형 전방 탐색 (?=...)
- nagative (부정)
  - 부정현 전방탐색 (?!...)

```python
# http, https 프로토콜 형식
# https://ID:PW@DOMAIN:PORT/PATH <- URL을 풀로 채운 형식
# https://DOMAIN/PATH <- URL을 약식으로 처리한 형식

# 여기서 https라는 프로토콜명을 추출하시오
text = 'https://m.naver.com' 
p = re.compile('.+:')
m = p.search( text )
if m:
  print( m.group() )
  # :이 같이 나와서 클린하지 않다
  # : 문자 앞쪽을 탐색
```

```python
# 긍정형 전방탐색
# 나름 조건이 어려운 상황에서 (정규직 적용이 까다롭다) 유용하게 사용
p = re.compile('.+(?=:)')
m = p.search( text )
if m:
  print( m.group() )
```

```python
# 부정형 탐색
# 파일들 중에서 확장자가 py, msi로 끝나는 파일은 제외한다
p = re.compile('.*[.](?!py$|msi$).*$')
```

```python
p.search('a.hwp').group()
```

```python
print( p.search('a.py') )
```

```python
print( p.search('a.msi') )
```

```python
print( p.search('a.py2') )
```

# 주민번호 필터링 

- 000000-0000000 => 000000-*******
- 000000-0000000 => 000000-1******

```python
# 주민번호에 대한 의미는 배제하고, 진행
# {7,7} => {7} 동일
p = re.compile( '(\d{6})[-]\d{7,7}' )
```

```python
data = '''
A 940101-1234567
B 990101-1234567
'''
```

```python
# \g<1> : 1번 그룹 참조
print( p.sub( '\g<1>-*******', data) )
```

```python
# 000000-0000000 => 000000-1**
# 정규식
p = re.compile( '(\d{6})[-](\d)\d{6}' )
# 결과확인
print( p.sub( '\g<1>-\g<2>******', data) )
```
