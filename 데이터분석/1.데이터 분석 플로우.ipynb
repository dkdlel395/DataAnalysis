  - 무료 서비스 -> 데이터 수집 목적이 존재

## 데이터 종류

- 데이터 종류
  - 정형
    - **구조화 된 데이터**
    - 대부분 출처가 **관계형데이터베이스**를 통해서 추출된 데이터 확률이 높다
    - 데이터와 구조정보(스키마)가 분리되어 있다
  - 비정형
    - 구조가 없는 데이터
    - 바이너리 데이터
      - 비디오, 오디오, 문서파일(doc,hwp..)
    - NoSQL
      - 빅데이터에서 다루는 형태
      - 로그, 접속기동등등...
  - 반정형
    - RDBMS(관계형)에서 추출(구성)하진 않았지만, 어느 정도 구조는 가지고 있다
    - JSON(키-값 형), XML -> 텍스트 기반,이기종,플랫폼, 언어(랭귀지) 독립적인 데이터 포멧 
    - HTML
    - 데이터안에 데이터+구조정보(스키마)가 같이 존재

## 데이터 수집 방식

- 데이터 수집 전략(난이도)
  - 모든 데이터는 그 사용에 관해서 법률적인 검토가 필요하다
  - level 1
    - 정형/반정형된 데이터가 대상(일반론)
    - 종류
      - 사내 데이터
      - 공공 데이터(국가기관, 민관, )
      - 경쟁 대회 데이터(캐글, 데이콘, 아레나, 지자체,..)
        - 기간이 지나면 미제공 할수 있음
      - 국책
        - AI허브(https://aihub.or.kr/)
      - 바우처
        - 국가에서 데이터 구매시 사용하라고 제공하는 구매비용제공쿠폰(신청후 심사후 제공)
        - 데이터 거래소(https://kdx.kr)
---


- 인터넷 통해서 직접적 수집(레벨 1에서는 내가 원하는 데이터를 얻을수 없었다)
  - level 2
    - OPEN api 활용
    - 특정기업의  공개용 API를 이용하여 수집
    - 일일 쿼터량 제한
    - 데이터는 반정형 데이터(출처는 정형 데이터) - xml or json

  - level 3
    - Web Scraping(웹스크래핑)
    - 웹사이트 접속후 어떤 액션도 없이 바로 로딩된 html을 긁어서, 파싱후 데이터 추출
    - 환율, 위키피디아, ...
    - BS4

  - level 4
    - Crawling(크롤링)
    - 사용자의 노력이 들어가서 특정 정보에 접근할수 있는 사이트가 대상
      - 로그인, 스크롤, 클릭, ajax,....
      - 다른 관점에서는 매크로라고 표현함
      - 과거 php, js, 현재 python으로 많이함
    - Selenium

## 자동화

- 데이터 수집 빈도/주기 등등에 따라 매번 사람이 수집할수는 없다 
- 자동화를 통해서 알아서 수집->정제->적제등등 하나의 플로우로 작동되게 구성
- 방법
  - 1. 파이썬으로 스크립트(목적성)를 작성
  - 2. 하나의 배치파일로 작동되게 환경 구성
  - 3. 자동화 방식 (os에 종속적 구성)  
    - window server    
      - **작업 스케줄러 **
    - linux
      - **cron**
    - 해당 서버는 통상 클라우드에 구성하여 처리(로컬 PC도 가능함)
- 주의
  - 수집 주기가 짧으면 => 디도스 공격으로 인식할수 있다(타켓사이트)
  - 주기를 적절하게 조정 혹은 비번해야 하면 프록시(IP 추척을 막기 위해)를 구성해서 우회하여 수집
    - 특정 IP에서 지속적으로 들어온다는 데이터를 숨기는 용도
  - User Agent라는 요청시 헤더값을 조정하여서(매변 변조하여서) 수집하면 좀더 안정적

# 모델 구축

- 통계 모델
  - 통계 기반
- 머신러닝(ML) 모델
  - 지도 학습(정답이 있다)
    - 분류 (이 사진은 고양이일 확률이 95%)
    - 회귀 (이 집은 xxxx원으로 예측된다,..->가격예측, 수치값, 범위가 존재)
  - 비지도 학습 (정답이 없다)
    - 군집 (클러스터링), 넷플릭스 고객 분류
    - 차원축소(압축), 전처리에 적용될 확률이 높음
  - 준비도 학습
    - 지도 + 비지도
  - 강화 학습
    - 어떤 행동에 따른 리워드(보상)이 주어지면 그 행동을 강화할려는 경향이 보인다 -> 학습에 적용 -> 학습시키고자 하는 대상(에이전트)이 최대 보상을 얻도록 행동을 강화하는 학습법 -> 유한게임 환경에서 구현이 가능(무한 게임은 쉽지 않다)
    - ml-ganet(시뮬레이션), 기타 대기업들 위주..
  - 머신러닝은 **알고리즘**을 훈련시켜서 사람이 할수 없는(해결못한는) 일을 해결하고자 함이 목표
    - 알고리즘, 하이퍼파라미터값

- 딥러닝(DL) 모델
  - 딥러닝 인간의 뉴런구조(신경구조)를 응용하여서 사람이 해결하지 못하는 특정 주제를 해결하기 위해 **인공신경망**을 구성하고 이를 학습을 통해서 해결하는것을 목표로함
    - 신경망 구조, 가중치(W), 편향(B)을 저장하고, 필요할때 가져다 사용하면 언제들지 사용 가능함
  - 영상, 이미지, 자연어, 생성, ....
  - CNN부터 파생, RNN부터 파생, GAN부터 파생....
  - **현실적으로 내가 만든모델이 기관/기업/연구소등등 에서 만든 모델보다 우수할것인가?**
    - **프로젝트 -> 전이학습 전략 선회**
    - **전이학습 -> 파인튜닝(신경망을 일부개선), 기존 가중치 + 새로운 파트 가중치학습 -> 모델완료** => 데이터 전처리, bm모델(미고려)
    - 기존 모델을 개선 -> 학습 -> 결과를낸다
      - 멘토별로 달라질수 있다
  

# 시스템 통합/신규서비스 구현

- 실세 구성한(계획한) 시스템(서비스)에 앞에서 구성한 모델을 적용하여 실제 서비스를 구축/운용/유지보수 하는 단계
- MLOps
  - 현재 적용되면서, 정립되고 있는 단계
  - 데이터 분석가, 모델러, DevOps, 등등 담당자가 MLOps로 업무중이다(현재는)
    - 데이터 분석가, 모델러 -> MLOps
    - 특정 포지션으로 취업후 내부에서 해당 업무로 업그레이드
- 구성 내용 중심으로 내용
  - 서비스(웹, flask, Django)
  - 데이터
    - 학습용 데이터
    - 검증용 데이터
    - 테스트용 데이터
    - 서비스중 발생하는 고객에 만드는 데이터(활동, 업로드, 서비스 형태 마다 다름)
  - 모델
    - 학습
      - 버전 관리
        - 가중치, 편향, 정확도 같은지표 관리(대시보드등 관리)
        - 모델 서빙할때 업그레이드, 다운로그레이드가 쉬워야 한다
  - 환경
    - 서비스가 구동되는 환경 Docker 기반 구축
    - docker에 모델이 세팅된 형태가 아닌 요청시 제공해 주는 형태로 구성
    - docker에서 쿠버네티스를 이용한 확장성 및 유연성을 보장(트레픽이 많으면 용량을 키우고,...)
  - 기타 환경
    - aws 서비스 환경
    - s3에서는 데이터 업로드, 관리등등 구축

- 고객에게 최적의 서비스를 제공, 관리자는 편하게 관리, 모델을 유연하게 서빙할 수 있는 환경 구축하는 포지션
  - 초기는 심플하게 구성
  - 사용자 늘어가고, 용량이 부적해지면 확장 한다
