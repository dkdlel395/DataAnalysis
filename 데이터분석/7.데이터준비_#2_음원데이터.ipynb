## 제작사 데이터 검토 및 정제

```python
music.Distribution
# 데이터가 띠어쓰기, 등등 문제로 같은 내용이 여러 버전으로 보인다
music.Distribution.unique(), len(music.Distribution.unique())
```

```python
# 데이터별로 몇번 등장했는지 체크
music.Distribution.value_counts()
```

```python
def myChange( x ):
  x = x.replace(' ', '')
  x = x.replace('㈜', '(주)')
  x = x.replace('소니뮤직', 'SonyMusic')
  x = x.replace('유니버설뮤직', 'Universal')
  x = x.replace('(주)지니뮤직', '지니뮤직')
  x = x.replace('(주)미러볼뮤직', 'MirrorballMusic')
  return x

tmp = music.Distribution.apply( myChange )
tmp.unique(), len(tmp.unique())
```

```python
music.Distribution = tmp
music.Distribution.value_counts()
```

```python
music.columns
```

```python
# Distribution 컬럼 기반으로 피벗 테이블 진행
pd.pivot_table( music, index=['Distribution'], 
               values=['Music Count', 'Rating'],
               aggfunc=[np.sum, np.mean, len])
# 피벗을 치면 불필요한 컬럼이 생길수도 있다 함수와 값들이 가각 n,m개가 되면
# n*m개 컬럼이 만들어지니까 => 개별 데이터별로 개별 함수를 적용하고 싶다 -> groupby
```

```python
# 컬럼별로 별도로 적용해서 -> 소트 정렬 -> 최대 평점순, 최대 제작순, 최다곡 제작순 -> 바차트로 이어짐
music.groupby('Distribution').agg({
    # 컬럼:함수
    'Rating':'mean',
    'Distribution':'count',
    'Music Count':'sum'
})
```

```python
music.groupby('Distribution').agg({
    # 컬럼:함수
    'Rating':'mean',
    'Distribution':'count',
    'Music Count':'sum'
}).sort_values( by='Rating', ascending=False)
# 제작(유통) 편수 최소 50건이상인 회사를 기준으로 집계 

# 소수 제작(유통)사가 음원 시장을 점유하고 있다
```
