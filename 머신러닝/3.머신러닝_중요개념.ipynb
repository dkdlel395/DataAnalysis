# 알고리즘 선택

## 지도학습 > 분류

### 로지스틱 회귀(*)

- 종속변수와 독립변수가 선형관계 있는 데이터
- 기본알고리즘, 이진 분류 상황에서 사용
- 베이스라인 알고리즘

### KNN-(최근접이웃)(*)

- 아웃라이어(이상치):다른 관측치에 비해 비정정삭적으로 동 떨어져 있는 값
- 이상치가 적은 데이터에 적합한 알고리즘
- 다중분류문제, 베이스라인 모델

### 나이브베이즈(*)

- 피처(특성)의 종류가 매우 많은 경우
- 딥러닝 제외, **자연어 처리시**
- 피처들이 모두 독립적이고, 범용성이 높지 않은 경우

### 결정트리(*)

- 예측몇 성능면 단독으로 사용 않함
- 트리기반 앙상블 알고리즘의 베이스모델

### SVM(서포트백터머신)-SVC

- 하나의 패턴 인식, 자료 분석을 위한 모델
- 이진 선형 분류 모델에 적합

### 앙상블

- 약한 분류기를 여러개 모아서 강한 분류를 만드는 기법

#### 보팅

- 다른 여러개의 알고리즘을 모아서 분류기 생성

#### 베깅-랜덤포레스트(*)

- 동일 알고리즘을 여려개 모아서, 보팅을 수행한다
- 트리기반 앙상블 기법중에 가장 일반적 모델
- 랜덤포레스트는 결정트리 n개를 모아서 만든 모델
- 부스팅보다 모든면에서 열세, 과도기적 모델

#### 부스팅 (현재 캐글우승자들이 가장 많이 사용하는 알고리즘)

- LightBGM or CatBoost > XGBoost
- 항상 이 서열이 맞지는 않지만 추세가 이동중이다

##### XGBoost(*)

- 캐글 대회 우승자가 가장 많이 사용한 모델

##### LighGBM(*)

- MS에서 만듬
- 성능면에서 XG를 상회하고 있음

##### CatBoost(서드파트)

- 오프소스그룹, Sklearn에는 미포함

#### 스태킹

- 2단게 모델로 구성
- 1차 모델 : 기본 데이터를 이용하여 예측 수행, 여러개의 서로 다른 알고리즘 사용
- 2차 모델 : 예측 결과를 데이터 이용하여, 이를 학습하여 다시 예측, 메타학습 기법 적용

## 지도학습 > 회귀

### 선형회귀(*)

- 종속변수와 독립변수간의 관계가 선형으로 구성된 데이터
- 예측 성능은 떨어짐, 베이스모델용도

### KNN-(최근접이웃)(*)

- 주변의 가장 가까운 K개의 샘플을 통해서 값을 예측하는 방식
- 베이스라인 모델

### Lasso(라쏘)

- 선형 회귀의 단점을 보와하기 위해 등장
- L1 정규화 - 패널티를 부여하는 기법

### Ridge(릿지)

- 선형 회귀의 단점을 보와하기 위해 등장
- L2 정규화  - 패널티를 부여하는 기법

### 결정트리(*)

### SVM(서포트백터머신)-SVR

### 앙상블 (이하생략)

## 비지도학습 > 군집

### K-Mean(*) (거리 기반)

- N개의 군집의 중심점을 특정한 임의의 지점을 선택하여서, 해당 중심점에 **가까운** 포인트들이 그중심점의 맴버들이 되고, 이 과정 반복하면서 군집화하는 전략

### Mean-Shift (밀도 기반)

- 특정한 대역폭(bandwidth)을 가지고, 밀도가 높은쪽으로 중심을 이동을 하는 방식

### GMM (확률 기반)

- 데이터가 여러개의 가우시안 분포(가우시안 정규분포)가 섞여서 데이터가 분포된것이다라고 가정하여서 군집을 수행하는 방식

### DBSCAN (밀도 기반, 기하학적 분포)

- 데이터의 분포가 기하학적이다
- 특이 케이스

## 비지도학습 > 차원축소

### PCA(*)

- 여러 변수간에 존재하는 상관관계를 이용하여서 이것들을 대표하는 **주성분을 추출**하여  차원을 축소하는 기법

### LDA

- 잠재 디클레어 할당 : 자연어, 문서상에 잠재되어 있는 토픽을 이용하여 문서를 분류한다(예)

### SVD

- PCA가 내부적으로 활용
- 고차원 피처 데이터를 2차 행렬로 분리하는 행렬 분해 기법

### NMF

- 고차원 피처 => 저차원 피처로 만드는 행렬 분해 기법
